```python

import pandas as pd
import os
import numpy as np
```

```python
filename1 = os.path.join(os.getcwd(), 'iot-datasets', 'resource_type.csv')
filename2 = os.path.join(os.getcwd(), 'iot-datasets', 'train.csv')
resource_df = pd.read_csv(filename1, header = 0)
train_df = pd.read_csv(filename2, header = 0)
```

```python
train_df.head()
```

```python
resource_df.head()
```

```python
#check for uniqueness of id column in both dataframes.
#if id column is unique in both dataframes, then it is one to one mapping.
#if id column is not unique in both dataframes, then it is one to many mapping.
res_unique = resource_df['id'].is_unique
train_unique = train_df['id'].is_unique
```

```python
#one to many = dataframe contains non-unique values which
#means there are multiple rows with same id.
#one to one = dataframe contains unique values
#combination of both dataframes will result in one to many mapping???

if res_unique and train_unique:
    print("One to One Relationship")
else:
    print("One to Many Relationship")
```

```python
#create a dictionary to store id and relationship
id_relationship = {}
#iterate through id column of resource_df
for id in resource_df['id']:
    if res_unique and train_unique:
        id_relationship[id] = 'One to One'
    else:
        id_relationship[id] = 'One to Many'
```

```python
#PRINT the relationship for each id
for id, relationship in id_relationship.items():
    print(f'id: {id} relationship: {relationship}')
```
